#  to use plotnine we have gog (grammer of graphics)
1. create canvas -> ggplot(dataset)
2. create axis  aes(x ="x-axis" , y="y-axis", color="#234234" , fill="colors of bars")
3. type of graph -> geom_bar() , geom_line(), geom_point()
4. facets -> it is used to divide our graph horizontally or vertically or both 
    facets_grid(<rowname>/. ~ <colname>/ . )
-> use "." if we do not want to pass the coloumn or row
