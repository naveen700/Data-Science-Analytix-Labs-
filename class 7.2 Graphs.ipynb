```python
import pandas as pd

from matplotlib  import pyplot as plt
import matplotlib as mp

%matplotlib inline

import seaborn as snb
```

```python
sales = pd.read_csv('../DataSets/1000 Sales Records.csv').head(500)

snb.lmplot(x="TotalCost", y="TotalProfit",data=sales, fit_reg=False, hue="Region")
# snb.boxplot(data=sales)
```

```python
# boxplot or whishker plot (both are same), it can take series or full dataframe two but it doesnot take categorical variable, if present in data frame it will be ignored automatically.
snb.boxplot(data=sales.OrderID)
```

```python
# Histogram  - using sns.displot()(distribution plot)
snb.distplot(sales.TotalProfit)
# curve represents the distribution curve or bell curve
# adding all the bar percentage or values(present in y axis) should add and give 1
```
