##### probability distribution
probablity distribution is a statistical function that can describes all possible values and likelihoods that a random variable can take within given range.

(my own definition)distribution -> for given range , what is the probability of particular set of observation will belong to that range.

1. PDF(probability density function) :- we can calculate the probability of observation data coming under same range using integration by calculating the area under curve.

problem:- to calculate integrtation  /f(x)d(x) we must know f(x), and there could be many types of possible f(x) because of diffrent possible curves of observed data , so how do we know which f(x) we should take?
Reasearcher have already created f(x) equation for most common curve patterns,which  we can see at web. 95% data follows these already defined curve patterns and we can use f(x) equation for that curve 
1.we can match the shape of our data curve to max matching already defined curves and decide that this curve's equation we can use.
eg:-data is following normal distribution curve so we have normal distribution equation(mathemetical form) for our data  already.

2.there is also statistical test by which we can know which probability distribution our data follows  eg:data follows  normal distribution , log normal distribution.poisson distribution ,chisquared distributin etc.

why we need to know the distribution?
ans. by knowing which type of distribution it follows , we can calculate the probability of set of observation falling into given range.
